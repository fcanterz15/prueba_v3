# 5.3 Redefinición de posiciones de los jugadores

Como primer objetivo del Trabajo Fin de Máster, fue definido la redefinición de las posiciones de los jugadores de baloncesto a partir de las diferentes métricas estadísticas que estos obtuviesen a lo largo de una temporada. Para ello, a través de este notebook, será definido el proceso de refinición de posiciones y el análisis de estas nuevas posiciones. Para ello, será implmentado el algoritmo de clusterización k-meams con el que se espera obtener las nuevas posiciones a través de las estadísticas obtenidas. Además, previsamente a la implementación del algoritmo k-means, será realizada una reducción de dimensiones del conjunto de datos mediante el método PCA para la mejora del rendimiento del algoritmo de agrupamiento.

Teniendo esto presente, este notebook, estará dividido en los siguiente apartados:

## 1. Reducción de dimensiones.

En este apartado, se procede a realizar la *reducción de dimensiones* que es una técnica empleada en análisis de datos y aprendizaje automático que permite disminuir el númer de variables en un cojunto de datos mientras se consera la mayor cantidad posible de información relevante. Como ha sido definido al principio de este notebook, ha sido dedicido realizar esta reducción de dimensiones mediante el algoritmo de Análisis de Componentes Principales o también conocido como PCA que es una ténica que permite la reducción de dimensiones mediante la realización de los siguientes pasos:
1. Estandarización de variables. Se realiza la estandarización de los datos para que cada variable contribuya de manera equitativa al análisis mediante el uso de la media y dividiendo por la desviación estándar cada una de las variables.
2. Cálculo de la matriz de covarianza. Se genera la matriz de covarianza en la que se mide cómo varian conjuntamente las diferentes variables del conjunto de datos.
3. Cálculo de los componentes principales y sus valores. Se calculan los vectores de componentes principales y los valores que estos toman.
4. Selección de los componentes principales. Se ordena cada uno de los componentes principales en función de los valores que estos tomas siendo los primeros componentes principales los que menor valor tienes.
5. Transformación de los datos. Por último, se realiza la proyección de los datos originales sobre los nuevos ejes definidos por los nuevos componentes principales.

Mediante la implementación de estos pasos, son obtenidos los componentes principales que permiten la reducción dimensional de variables del conjuntos de datos principal. 

Una vez explicado de manera teórica el algoritmo, se pasa a realizar su implementación. Para ello, en primer lugar se realiza la carga del *Tablón de estadísticas* que será utilizado durante este notebook:

```python
# Ruta del tablón y carga del tablón
ruta_tablon = 'E:/TFM/02. Datos/ACB/TABLON ESTADÍSTICAS'
df_tablon_estadisticas = pd.read_csv(ruta_tablon + '/TABLON_ESTADISTICAS_FINAL.csv', sep = ';')

# Se realiza la extracción de volumetrías
print("#######################################")
print("# RESUMEN DEL TABLÓN DE ESTADISTICAS")
print("#######################################")
print("Nº de registros:",df_tablon_estadisticas.shape[0])
print("Nº de columnas:",df_tablon_estadisticas.shape[1])
print("Columnas:", (' | ').join(df_tablon_estadisticas.columns))
print("Nº de elementos vacíos:",(df_tablon_estadisticas=='').sum().sum())
print("Nº de elementos nulos:",df_tablon_estadisticas[df_tablon_estadisticas.columns].isna().sum().sum())
print("\n",df_tablon_estadisticas.head().to_string(index=False, col_space=4))
```

Tras la carga de los datos y previo paso a la implementación del algoritmo *PCA* se pasa a realizar la selección de las variables numéricas del conjunto de datos y a la normalización de las variables que representan acciones del juego como son el ORB%, DRB%, STL%, BLK%, TOV%, AST% y FTr mediante el método Min-Max que es el método que realiza la normalización de los datos a un rango específico entre [0,1]. El motivo de la utilización de esta normalización es para poder conseguir que todas las variables tenga la misma escala y en las variables en la que se realiza la normalización los jugadores que hayan tengan mejores número en este apartado se vean recompensados durante la reducción de dimensiones:

```python
# Se realiza la normalización de las variables estadísticas
scaler = MinMaxScaler()
df_tablon_estadisticas_aux = df_tablon_estadisticas[['ORB%', 'DRB%', 'STL%','BLK%', 'TOV%', 'AST%', 'FTr', 'TL%', 'PINTURA_p', 'PINTURA_v','MID_RANGE_p', 'MID_RANGE_v', '3FRONT_p', '3FRONT_v', '3CORNER_p','3CORNER_v']]
df_tablon_estadisticas_aux[['ORB%','DRB%', 'STL%', 'BLK%', 'TOV%', 'AST%','FTr']] = scaler.fit_transform(df_tablon_estadisticas_aux[['ORB%','DRB%', 'STL%', 'BLK%', 'TOV%', 'AST%','FTr']])

print("#######################################")
print("# TABLÓN DE ESTADISTICAS NORMALIZADO")
print("#######################################")
print(df_tablon_estadisticas_aux.head().to_string(index=False, col_space=4))
```

Tras la selección de las variables numéricas y la normalización de las estadísticas de acciones de los jugadores, se pasa a la aplicación del algoritmo PCA sobre el conjunto de datos. Para su entendimiento, junto a la aplicación del algoritmo se pasa a visualizar la varianza de cada uno de los componentes principales:

```python
# Inicializa el objeto PCA y ajuste del PCA a los datos
pca = PCA()
pca.fit(df_tablon_estadisticas_aux)
explained_variance_ratio = pca.explained_variance_ratio_
cumulative_explained_variance = np.cumsum(explained_variance_ratio)
componentes = np.arange(1, len(explained_variance_ratio) + 1)

# Creación de la visualización
fig, ax = plt.subplots(figsize=(20, 6))
line_exp_var = ax.plot(componentes, explained_variance_ratio, marker='o', linestyle='-', label='Varianza Explicada')
line_cum_var = ax.plot(componentes, cumulative_explained_variance, marker='o', linestyle='-', label='Varianza Acumulada')
ax.set_yticks(np.arange(0, 1.1, 0.1))
ax.set_xticks(componentes)
ax.set_xlabel('Componente')
ax.set_ylabel('Varianza Explicada / Acumulada')
ax.set_title('Varianza Explicada y Acumulada por Componente Principal')
ax.legend()

for x, y in zip(componentes, explained_variance_ratio):
    ax.text(x, y, f'{y:.2f}', ha='right', va='bottom')
for x, y in zip(componentes, cumulative_explained_variance):
    ax.text(x, y, f'{y:.2f}', ha='right', va='bottom')

plt.show()
```

Tras la aplicación del algoritmo PCA y la creación de la visualización, se ve claramente para cada uno de los componentes su grado de explicabilidad en la varianzada del conjunto de datos original. Debido a que se ha definido como obtimo un umbral superior 85% en la explicabilidad de los componentes a seleccionar, ha sido decidido seleccionar el número de componentes sobre los que realizar la reducción dimensional en **8** puediendo así reducir las dimensiones del conjuntos de datos a la mitad. 

Por lo tanto, se pasa a realizar la reducción dimensional del conjunto de datos en 8 componente principales y se pasa a crear el conjunto de datos que será utilizado para la obtención de las agrupaciones:

```python
# Inicializa PCA y ajusta a tus datos a 8 componente principales
pca_8D = PCA(n_components=8) 
pca_8D_data = pca_8D.fit_transform(df_tablon_estadisticas_aux)
```

## 2. Obtenición de agrupaciones

Tras la reducción de dimensiones del conjunto de datos original, el siguiente paso a realizar es la obtención de agrupaciones en las que serán distribuidos los jugadores de baloncesto y que darán lugar a las nuevas posiciones de baloncesto. Para ello, este apartado estará dividido en dos secciones diferentes:
* *Selección del número óptimo de agrupaciones.* Se realizará la selección del número óptimo de agrupaciones en las que será distribuidos los datos.
* *Aplicación del algoritmo kmeans.* Se implementará el algoritmo k-means sobre el conjunto de datos reducido dimensionalmente para obtener las agrupaciones a las que pertenece cada uno de los jugadores.

### 2.1 Selección del número óptimo de agrupaciones.

La selección del número óptimo de agrupaciones, o clusters, es la etapa dentro de la clusterización en la que se determina cual es el número de agrupaciones se ajusta mejor al conjunto de datos sobre el que se quiere aplicar el algortimo de clusterización. No existe ningún método que permita determinar el número de clusters debido a que la selección del número óptimo de clusters depende de la naturaleza de los datos y el objetivo del análisis. Es por este motivo, que para la obtención del número óptimo de clusters ha sido decidido que se utilice el *Criterio de la silueta* que es un método que permite media la calidad de los clusters a partir de los similares que son los puntos dentro de un cluster con respecto a otros puntos dentro de otro cluster. Es por este motivo, que ha sido realizado el cálculo de la puntuación de la silueta para diferentes número de clusters y la visualización de cada uno de esas siluetas:

```python
# Se define número de agrupaciones a observar y se generan las visaulizaciones
range_n_clusters = [2, 3, 4, 5, 6, 7, 8]
for n_clusters in range_n_clusters:
    obtener_silueta(pca_8D_data, n_clusters)
```

Tras la extración de las siluetas para los valores de 2 a 8 clusters, se puede observar como la división del conjunto de datos en 5, 7 y 8 clusters no son correctas debido a su alto porcentjae de valores negativos además de que la distribución dentro de que existen siluetas para estos números de clusters que no superar el valor promedio de la silueta. Por otro lado, el número de agrupaciones de 2. 3 y 4 si que dan un porcentaje de error asumible los grupos que se generan son demasiado grandes y se aprecia que existen agrupaciones posibles dentro de esos cluster. Por lo tanto, el número de cluster óptimo que ha sido seleccionado es el de **6** debido a que el porcentaje de error es de 0.15 que es un valor asumible y todos los clusters tiene una distribución forma muy simétrica.

### 2.2 Aplicación del algoritmo kmeans

Tras la obtención del número óptimo de clusters, se pasa a realizar la implementación del algoritmo de clusterización. Com ha sido definido anteriormente, se ha obtado por implemental el algoritmo k-means debido a su eficiencia computacional para el trabajo con conjuntos de datos y la facilidad a la hora implemtar que tiene el algoritmo. Con respecto al funcionamiento del algoritmo k-means este puede dividirse en los siguiente pasos:
1. *Inicialización.* Selección de k puntos aleatorios del conjunto de datos como los centroides iniciales. El valor de k es dado al algoritmo debido a que es un algoritmo supervisado.
2. *Asignación.* Asigna cada punto de datos al centroide más cercano.
3. *Actualización^.* Recalcula los centroides de los k clusters utilizando la media de todos los puntos asignados al cluster.
4. *Repetición.* Se repiten los paso 2 y 3 hasta que no se produzcan cambios significativos o se alcance el número maximo de iteraciones.

Tras definir el algoritmo se pasa a su inicialización y entrenamiento para 6 clústers y 8 componentes:

```python
# Se inicializa el algoritmo k-means y se realiza su entrenamiento
algo_kmeans = KMeans(n_clusters=6, random_state=42)
algo_kmeans.fit(pca_8D_data)
```

Como puede observarse la implementación del algoritmo es sencilla. Adicionalmente a este paso, se va a incluir el cluster al que pertenece cada uno de los jugadores:

```python
# Se asigna el cluster obtenido
df_tablon_estadisticas['CLUSTER'] = algo_kmeans.labels_
```

## 3. Análisis de las nuevas posiciones

Tras la reducción dimensional y la aplicación del algoritmo k-means, han sido agrupados los jugadores por sus estadísticas a lo largo de las temporadas. Es por este motivo que durante este apartado se realizará el análisis de los resultados tras la clusterización. Como primer paso, se pasa a mostrar la distribución de los jugadores dentro de cada grupo:

```python
# Obtención de volumetrías por cluster
vol_clusters = df_tablon_estadisticas['CLUSTER'].value_counts()
vol_clusters = vol_clusters.sort_index()

# Creación de la visualización
plt.figure(figsize=(10, 6))
plt.axis('equal')
wedges, texts, autotexts = plt.pie(vol_clusters, autopct=lambda pct: f"{pct:.1f}%\n{int(pct/100.*sum(vol_clusters))}", startangle=90, wedgeprops=dict(edgecolor='black'))
for text in texts:
    text.set_color('black')
for autotext in autotexts:
    autotext.set_color('black')


plt.title('Distribución de clusters')
plt.legend(wedges, vol_clusters.index, loc="center left", bbox_to_anchor=(1, 0, 0.5, 1), title='Clusters', labels=vol_clusters.index)
plt.show()
```

Como puede observarse, dentro de cada uno de los grupos, el número de jugadores asignados a esto es similar siendo el cluster en el que hay más presencia el número 1 mientra en los que mnos son el 3 y el 4. Se pasa ahora al análsis de cada uno de los clusters con el fin de identificar el tipo de perfil de jugadores que componen los cluster. Para ello, se realiza la creación del tablón de medias del *Tablón de estadístas* y se carga la información referencia a los jugadores en la tabal de control:

```python
# Carga de información de la tabla de control
ruta_estadisticas = 'E:/TFM/02. Datos/ACB/ESTADISTICAS'
df_tablon_control = pd.DataFrame()

# Se extrae el listado de directorios de temporadas
l_dir_temporadas = ['2020', '2021', '2022']

for temporada in l_dir_temporadas:
    # Se extrae el listado de directorios de equipos
    ruta_dir_temporada = ruta_estadisticas + '/' + temporada
    l_dir_equipos = os.listdir(ruta_dir_temporada)
    
    for equipo in l_dir_equipos:
        # Se inicializa la ruta de estadísticas de un equipo
        ruta_dir_estadisticas_equipo = ruta_dir_temporada + '/' + equipo
        
        # Se realiza la carga de ficheros de estadísticas de jugadores
        df_tablon_control_equipo = pd.read_csv(ruta_dir_estadisticas_equipo + '/05_INFORMACION_JUGADORES.csv', sep = ';')
        df_tablon_control_equipo['TEMPORADA'] = int(temporada)
        df_tablon_control_equipo['EQUIPO'] = equipo
        df_tablon_control = pd.concat([df_tablon_control,df_tablon_control_equipo])
        
df_info_cluster = df_tablon_estadisticas[['TEMPORADA','EQUIPO','DORSAL','JUGADOR_BS','CLUSTER']]
df_tablon_control = df_tablon_control.merge(df_info_cluster, left_on=['TEMPORADA','EQUIPO','DORSAL','JUGADOR_BS'],right_on=['TEMPORADA','EQUIPO','DORSAL','JUGADOR_BS'], how='inner')        

# Creación del tablón de medias
columnas = df_tablon_estadisticas_aux.columns
media_tablon = df_tablon_estadisticas[columnas].mean()
std_tablo = df_tablon_estadisticas[columnas].std()
df_medidas = pd.DataFrame({'Media': media_tablon, 'Desviación':std_tablo}).transpose().reset_index().rename(columns={'index':'MEDIDA'})
```

Tras las obtenición de estas dos tablas, se pasa a realizar el análisis de cada uno de los clusters obtenidos con el fin de obtener que perfil de jugadores conforman cada uno de ellos:
