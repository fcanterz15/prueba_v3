## 3. Análisis de correlaciones

Tras el análisis univariante de cada una de las variables que conforman el *Tablón de estadísticas* se pasa a realizar el estudio de correlaciones con el fin de ver como se relacionan las variables entre sí:

```python
# Selección de variables numéricas con las que evaluar la correlacion entre estas
df_corr = df_tablon_estadisticas[['ORB%','DRB%', 'STL%', 'BLK%', 'TOV%', 'AST%','FTr','TL%', 'PINTURA_p', 'PINTURA_v', 'MID_RANGE_p', 'MID_RANGE_v','3FRONT_p', '3FRONT_v', '3CORNER_p', '3CORNER_v']]
corr = df_corr.corr()

# Se genera la visualizacion
sns.set(style="white")
plt.figure(figsize=(20, 8))
sns.heatmap(corr, mask=np.triu(np.ones_like(corr, dtype=bool)), fmt=".1f", cmap='coolwarm', linewidths=.5, vmin=-1, vmax=1, annot=True)
plt.title('Mapa de correlación de variables')
plt.show()
```

Como puede observarse, las correlaciones se encuentra estabilizadas y son asumibles. Sin embargo hay que destacar que la alta correlación entre la variable *ORB%* y las variables *DRB%*, *BLK%* y *PINTURA_v* se debe a que los jugadores interiores se caracterizan por la relación de estas tres variables por lo que es asumible.
